```python
from datasets import load_dataset

raw_dataset = load_dataset("webis/tldr-17", split="train", streaming=True)
# only get 'subreddit', 'id', 'content', 'summary'
dataset = raw_dataset.remove_columns(['author', 'body', 'normalizedBody', 'subreddit_id'])
dataset = dataset.filter(lambda example: len(str(example['summary']).split()) >= 10)
```

```python
raw_dataset.dataset_size
```

```python
size = 0

data_iter = iter(dataset)

while (next(data_iter, None) is not None):
  size += 1
  if (size % 100000 == 0):
    size += 1
    print(size)
    print(next(data_iter, None))

print(size)
```
